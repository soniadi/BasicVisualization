## Library untuk visualisasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('seaborn')
```

## Bar Plot

```python
tipe=data["Tipe"].value_counts()
tipe
```

Feature tipe memiliki tipe data object. Sebelum melakukan bar plot, ubah terlebih dahulu sehingga didapatkan masing-masing feature pada tipe memiliki index tertentu berupa suatu angka.

```python
tipe.plot(kind="bar")
```

```python
tipe.plot(kind="barh")
```

```python
sns.countplot(y='Tipe',data=data,order = data['Tipe'].value_counts().index)
```

Dari hasil diatas didapatkan bahwa penjualan terbanyak berdasarkan tipe yaitu New Vixion Advance dan penjualan terendah adalah Force Elegan.

## Pie Plot

```python
data['Sub Kategori'].value_counts().plot(kind="pie")
```

Dari hasil diatas didapatkan bahwa penjualan terbanyak berdasarkan sub kategori yaitu Naked dan penjualan terendah adalah Maxi.

## Histogram

```python
data['Usia'].plot(kind='hist')
```

Dari hasil diatas didapatkan bahwa penjualan terbanyak memiliki range usia antara 25 sampai 34 tahun. Dari hasil tersebut ternyata usia 25 sampai 34 tahun merupakan usia produktif.

## Distribusi Plot

```python
sns.distplot(data["Usia"])
```

Hasil diatas menggambarkan bahwa penjualan sepeda motor terbanyak memiliki range usia antara 25 sampai 35 tahun dam mempunyai persebaran diantara range tersebut. Dari hal tersebut ternyata usia pembeli sepeda motor memiliki kecenderungan tidak berdistribusi normal.

## Boxplot

```python
data['Usia'].plot(kind='box')
```

Dari hasil diatas didapatkan bahwa penjualan terbanyak memiliki range usia antara 31 sampai 35 tahun. Hasil tersebut berbeda dengan menggunakan histogram, karena bosplot menggunakan konsep quartile sehingga pada boxplot lebih sensitif dalam melihat anomali atau sering disebut outlier. Namun pada kasus ini ternyata tidak terdapat usia yang outlier pada penjualan sepeda motor tersebut.

## Violin Plot

```python
sns.violinplot(y='Usia',data=data)
```

Dari hasil diatas didapatkan bahwa penjualan terbanyak memiliki range usia antara 31 sampai 35 tahun. Hasil tersebut berbeda dengan menggunakan histogram, karena bosplot menggunakan konsep quartile sehingga pada boxplot lebih sensitif dalam melihat anomali atau sering disebut outlier. Namun pada kasus ini ternyata tidak terdapat usia yang outlier pada penjualan sepeda motor tersebut. Violin plot juga menggambarkan persebarannya melalui gabungan distribusi plot dan box plot

## Joint Plot 

```python
sns.jointplot('Usia','Harga Jual',data=data,scatter_kws=dict(color='c'),kind='reg')
```

```python
sns.heatmap(data[['Usia','Harga Jual']].corr(),linewidths=.5,annot=True,cmap='RdYlGn')
```

Hasil diatas dapat memberikan gambaran hubungan antara harga jual dan usia yang dapat menggambarkan bahwa usia dan harga jual sepeda motor tidak memiliki hubungan satu sama lain, walaupun pada joint plot menunjukkan terdapat hubungan namun korelasi heatmap menggambarkan bahwa tidak terdapat hubungan antara usia dan harga jual sepeda motor
